```python
import pandas as pd

url= "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url)
```

1 - Realizar a média da metragem para cada um dos bairros.

```python
media_bairros = dados.groupby('Bairro')['Metragem'].mean()
media_bairros
```

2 - Duas formas de selecionar os dados por bairro

```python
#primeira forma
dados.query('Bairro == "Vila Madalena"')
```

```python
#segunda forma
bairro = "Vila Madalena"
dados[dados['Bairro'] == bairro]
```

3 - Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, assim como tentar colocar alguma conclusão.

```python
numeroBanheiros = dados["Banheiros"].value_counts()
numeroBanheiros.head(10).plot.bar()
#Conclusão: Quantidade de moradias com quantidade de banheiros
```

```python
vagasBairro = dados.groupby("Vagas")["Vagas"].count()
vagasBairro.plot.line()
#Conclusão: numero de vagas por bairro
```

4 - Pegar outras estatísticas dos dados (como média, mediana, mim, max).

```python
max = dados['Quartos'].max()        #maximo
min = dados['Quartos'].min()        #minimo
media = dados['Quartos'].mean()     #media
mediana = dados['Quartos'].median() #mediana

print('Numero maximo de quartos é: ', max)
print('Numero minimo de quartos é: ', min)
print('A media de quartos é: ', media)
print('A mediana de quartos é: ', mediana)
```
